```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace, OpenAIChatCompletionsModel
import os
from IPython.display import Markdown, display
from openai import AsyncOpenAI
```

```python
load_dotenv(override=True)
```

```python
ANTHROPIC_BASE_URL = "https://api.anthropic.com/v1/"

anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')

anthropic_client = AsyncOpenAI(base_url=ANTHROPIC_BASE_URL, api_key=anthropic_api_key)

anthropic_model = OpenAIChatCompletionsModel(model="claude-3-7-sonnet-latest", openai_client=anthropic_client)
```

```python
physicist1 = "You are Niels Bohr and you love discussing quantum mechanics."
physicist2 = "You are Richard Feynman and you love discussing quantum mechanics."


f1_agent = Agent(
    name="Niels Bohr",
    instructions=physicist1,
    model = "gpt-4o-mini"
)
f2_agent = Agent(
    name="Richard Feynman",
    instructions=physicist2,
    model = "gpt-4o-mini"
)
```

```python
tool_description = "Answer questions about quantum mechanics to a primary school student."

tool1 = f1_agent.as_tool(tool_name="Niels_Bohr", tool_description=tool_description)
tool2 = f2_agent.as_tool(tool_name="Richard_Feynman", tool_description=tool_description)

tools = [tool1, tool2]

tools
```

```python
physicist3_instructions = """You are Albert Einstein who don't like quantum mechanics concepts.
Use the tools provided to you to answer questions about quantum mechanics.
Decide which tool Niels_Bohr or Richard_Feynman is best suited to answer the question and use it.
Provide in the output which tool you used to get the answer.
Provide clear and concise explanations suitable for a primary school student.
"""

physicist3_agent = Agent(
    name="Albert Einstein",
    instructions=physicist3_instructions,
    model = anthropic_model,
    tools=tools)

message = "Please explain qantum interpolation and it's consequnces and applications."

with trace("Albert Einstein using tools"):
    result = await Runner.run(physicist3_agent, message)

# Format the output
formatted_output = f"Albert Einstein's Response:\n\n{result.final_output}"

```

```python
display(Markdown(formatted_output))
```
